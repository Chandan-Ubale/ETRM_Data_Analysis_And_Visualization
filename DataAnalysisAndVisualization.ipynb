a) Load all six data formats (CSV, JSON, Excel, TXT, HTML, XML) into Pandas 
DataFrames.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

etrm_trades_csv = pd.read_csv("etrm_trades.csv")
etrm_trades_html = pd.read_html("etrm_trades.html")[0]   # first table
etrm_trades_json = pd.read_json("etrm_trades.json")
etrm_trades_txt = pd.read_csv("etrm_trades.txt", delimiter="|")  # if pipe-delimited
etrm_trades_xlsx = pd.read_excel("etrm_trades.xlsx")
etrm_trades_xml = pd.read_xml("etrm_trades.xml")

print("CSV File:")
display(etrm_trades_csv.head())

print("HTML File:")
display(etrm_trades_html.head())

print("JSON File:")
display(etrm_trades_json.head())

print("Text File (pipe-delimited):")
display(etrm_trades_txt.head())

print("Excel File:")
display(etrm_trades_xlsx.head())

print("XML File:")
display(etrm_trades_xml.head())
```

b)Ensure consistency across datasets by verifying column names and data 
types.

```python
datasets = {
    "CSV": etrm_trades_csv,
    "HTML": etrm_trades_html,
    "JSON": etrm_trades_json,
    "TXT": etrm_trades_txt,
    "Excel": etrm_trades_xlsx,
    "XML": etrm_trades_xml
}

print("\n--- Column Names in Each Dataset ---")
for name, df in datasets.items():
    print(f"{name} Columns: {list(df.columns)}")

print("\n--- Data Types in Each Dataset ---")
for name, df in datasets.items():
    print(f"\n{name} dtypes:")
    print(df.dtypes)
```

a. Convert DeliveryStart and DeliveryEnd columns into datetime. 

```python
print("\n--Converting DeliveryStart and DeliveryEnd columns into datetime---")
for df in datasets.values():
    df["DeliveryStart"] = pd.to_datetime(pd.to_datetime(df['DeliveryStart'], errors='coerce'))
    df["DeliveryEnd"] = pd.to_datetime(pd.to_datetime(df['DeliveryEnd'], errors='coerce'))

print("\n--- After conversion Data Types in Each Dataset ---")
for name, df in datasets.items():
    print(f"\n{name}:")
    print(df[['DeliveryStart', 'DeliveryEnd']].dtypes)
```

Exploratory Data Analysis (EDA) 

a. Average price per commodity. -----EDA

```python
avg_price_per_commodity = etrm_trades_csv.groupby("Commodity")["Price"].mean()
avg_price_per_commodity
```

b. average price per commodity grouped by currency ----This EDA is not mentioned in the doc i just tried

```python
avg_price = etrm_trades_xlsx.groupby(['Commodity', 'Currency'])['Price'].mean().reset_index()
avg_price.rename(columns={'Price': 'AveragePrice'}, inplace=True)
avg_price
```

c. Distribution of trades by currency. -----EDA

```python
trades_by_currency = etrm_trades_csv['Currency'].value_counts()
print(trades_by_currency)
```

d. Trade periodicity breakdown (Daily, Weekly, Monthly, etc.).  -----EDA

```python
periodicity_breakdown = etrm_trades_csv['Periodicity'].value_counts()
print(periodicity_breakdown)
```

Bar chart showing total trade volume aggregated by each TradeType

```python
volume_by_trade_type = etrm_trades_csv.groupby('TradeType')['Volume'].sum().sort_values(ascending=False)
volume_by_trade_type.plot(kind='bar', color='lightgreen')
plt.title('Total Trade Volume by Trade Type')
plt.xlabel('Trade Type')
plt.ylabel('Total Volume')
plt.show()
```

Pie chart: Trades by Currency

```python
currency_counts = etrm_trades_csv['Currency'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(currency_counts, labels=currency_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Trades by Currency')
plt.show()
```

 Line chart: Average Price Trend by Delivery Start Date

```python
etrm_trades_csv['DeliveryStartDate'] = etrm_trades_csv['DeliveryStart'].dt.to_period('D')
avg_price_trend = etrm_trades_csv.groupby('DeliveryStartDate')['Price'].mean()
plt.figure(figsize=(10,5))
avg_price_trend.plot(marker='o')
plt.title('Average Price Trend by Delivery Start Date')
plt.xlabel('Delivery Start Date')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Histogram: Distribution of Notional Values

```python
etrm_trades_csv['NotionalValue'] = etrm_trades_csv['Volume'] * etrm_trades_csv['Price']
plt.figure(figsize=(8,5))
sns.histplot(etrm_trades_csv['NotionalValue'], bins=30, kde=True)
plt.title('Distribution of Notional Values')
plt.xlabel('Notional Value')
plt.ylabel('Frequency')
plt.show()
```

Heatmap: Commodity vs. Trader trade counts. here the trader column does'nt exist so i used an alternative for this commodity vs trade type like to see which commodity is most bought vs sold

```python
heatmap_data = etrm_trades_csv.pivot_table(index='Commodity', columns='TradeType', values='TradeID', aggfunc='count', fill_value=0)
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlOrRd')
plt.title('Commodity vs Trade Type Counts (Alternative to Trader)')
plt.show()
```
